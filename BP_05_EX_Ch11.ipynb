```python
from google.colab import files      #파일을 코랩에 업로드
uploaded = files.upload()       
filename = input("파일 이름을 입력하세오: ") #파일이름을 입력
infile = open(filename,"r")
read_file = infile.readlines()

outfile = open(filename, "w")
del_word = input("삭제할 문자열을 입력하시오: ")#삭제할 문자열 입력

for line in read_file: #반복문을 사용해 한줄씩 처리
    l, r = 0, len(del_word)
    new_line = "" #삭제하지 않는 문자열 저장
    while l < len(line): #반복문으로 삭제할 단어 검색
        if line[l:r] == del_word: #삭제할 단어와 찾은단어가 일치할 경우
            l += len(del_word)
            r += len(del_word)
        else: #삭제할 단어가 없을경우 더해주고 왼쪽변수에 저장
            new_line += line[l]
            l+=1
            r+=1
    print(new_line, file = outfile, end="") #new_line에 저장한 문자열을 outfile에 작성해줌

infile.close()
outfile.close()
print("변경된 파일이 저장되었습니다.")
```

```python
from google.colab import files      #파일을 코랩에 업로드
uploaded = files.upload() 
infilename = input("입력 파일 이름: ") #입력 파일 경로 입력
outfilename = input("출력 파일 이름: ") #작성할 파일 경로 입력

infile = open(infilename,"r")
outfile = open(outfilename,"w")
ans = 0 # 숫자를 더해줄 변수를 만들었음
k = 0 # infile에 있는 숫자가 몇 개나 있는지 확인하는 변수를 만들었음

for line in infile:
    ans += float(line) #숫자를 더해줌
    k += 1 #평균을 구할때 사용할 k값을 생성

print("합계="+str(ans), file = outfile) #output에 합계를 적어줌
print("평균="+str(ans/k), file = outfile) ##output에 평균을 적어줌

infile.close()
outfile.close()
```
